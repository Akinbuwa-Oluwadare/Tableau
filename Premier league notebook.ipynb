```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("premier_league_2019.csv")
```

```python
df.isna().sum()
```

```python
# Getting the most relevant columns for my anaylsis.
columns = ['date','matchDay','matchtime','round','teamId','pts','tot_points','wins',
           'draws', 'loses','scored','tot_goal','tot_con','Referee.x']
```

```python
df_copy = df.copy()
```

```python
df_copy = df_copy[columns]
```

```python
df_copy.head()
```

```python
#Splitting the date format into date and time seperately and assigning them to a new columns.
df_copy['Date'] = pd.to_datetime(df_copy['date']).dt.date
df_copy['Match_time'] = pd.to_datetime(df_copy['date']).dt.time
```

```python
df_copy.drop(["date","matchtime",], axis = 1, inplace = True)
```

```python
df_copy.head()
```

```python
#Renaming the columns names to a suitable one 
df_copy.rename(columns = {"matchDay": "Match_day",
             "round": "Round",
             "teamId": "Teams",
              "pts": "Points",
              "tot_points": "Total_points",
              "wins": "Wins","loses": "Loses",
              "tot_goal": "Total_goals",
              "tot_con": "Total_conceeded",
              "Referee.x": "Referees",
              "draws": "Draws","scored": "Goals_scored"},inplace = True)
```

```python
df_copy.head()
```

```python
# Re-arranging the columns 
re_arrange = ['Date','Match_day', 'Match_time', 'Round','Teams', 'Points', 'Total_points', 'Wins',
       'Draws', 'Loses', 'Goals_scored', 'Total_goals', 'Total_conceeded',
       'Referees']
df_copy = df_copy[re_arrange]
```

```python
df_copy.head()
```

```python
df_copy.to_csv("Premier_league_2019-2020.csv")
```
